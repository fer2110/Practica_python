```python
ruta_archivo = "C:\\Users\\mfer-\\Documents\\ChatLucia\\ChatLucia.txt"
with open(ruta_archivo, 'r', encoding='utf-8') as file:
    texto= file.read()
```

```python
import pandas as pd
import re


pattern = r'(\d{1,2}\/\d{1,2}\/\d{4}), (\d{1,2}:\d{2}\s[ap]\.\s?m\.) - (.*?): (.*)'

# Crear una lista de tuplas con los componentes extraÃ­dos
matches = re.findall(pattern, texto)

# Crear un DataFrame a partir de la lista de tuplas
df = pd.DataFrame(matches, columns=['Fecha','Hora', 'Usuario', 'Mensaje'])

# Mostrar el DataFrame
print(df)
```
